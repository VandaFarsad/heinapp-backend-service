```python
import os 
import django

os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'conf.settings')

django.setup()
```

```python
from calendar_app.services import CalendarService
from calendar_app.caldav.client import CalDAVClient
from datetime import datetime, timedelta, timezone


CALDAV_URL = "https://nextcloud.heina.org/remote.php/dav"
CALDAV_USER = "it@heina.org"
CALDAV_PASSWORD = "vhsqs2HYCpR8eva"
CALDAV_CALENDAR_URL = "https://nextcloud.heina.org/remote.php/dav/calendars/service/personal/"

CALDAV_CONFIG = {
    "url": CALDAV_URL,
    "username": CALDAV_USER,
    "password": CALDAV_PASSWORD,
    "calendar_url": CALDAV_CALENDAR_URL,
}


# Zielzeitfenster definieren
target_start = datetime(2026, 4, 25, 14, 0, tzinfo=timezone.utc)
target_end = target_start + timedelta(hours=8)

with CalDAVClient(config=CALDAV_CONFIG) as caldav_client:
    caldav_service = CalendarService(caldav_client=caldav_client)
    caldav_events = caldav_service.get_events(target_start, target_end)

# Ergebnisse anzeigen
print(f"Events im Zeitfenster {target_start} - {target_end}: {len(caldav_events)}")
for event in caldav_events:
    print(f"\n  Titel: {event.title}")
    print(f"  Start: {event.start_date}")
    print(f"  Ende: {event.end_date}")
    print(f"  Location: {event.location}")
    if event.rrule:
        print(f"  RRULE: {event.rrule}")
    if event.exdate:
        print(f"  EXDATE: {event.exdate}")
```
